### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
### The subplot() function is used to create multiple plots in a single figure. 
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### import numpy as np
#### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#### profit = np.array([3000, 8000, 1000, 10000])

```python
### A bar plot is used to represent categorical data with rectangular bars. It's useful for comparing different categories.
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(12, 5))

# Vertical bar plot
plt.subplot(1, 2, 1)
plt.bar(company, profit)
plt.title('Vertical Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
#### box1 = np.random.normal(100, 10, 200)
#### box2 = np.random.normal(90, 20, 200)

```python
### A box plot (also known as a box-and-whisker plot) is used to show the distribution of a dataset. 
### It's particularly useful for identifying outliers and comparing distributions.

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(10, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
